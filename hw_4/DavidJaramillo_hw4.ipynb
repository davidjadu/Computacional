#Gravedad en el planeta Brahe-3141-f
En este trabajo se analizarán 1000 series de datos que describen una movimiento parabólico en el planeta Brahe-3141-f, para esto se hará un ajuste de mínimos cuadrados, con el modelo de movimiento parabólico usual, para buscar las constantes $y_0$, $v_0$, $g$.
$$
y=y_0+v_0t+\frac{1}{2}gt^2
$$
Es importante hacer notar, que no todas las series se consideran válidas ya que algunas solo tienen ruido, por tanto, antes de hacer el ajuste de mínimos cuadrados se verifica, que los datos seán válidos. El proceso de verificación se encuentra explicado más abajo.
Además ya que cada serie de datos cuenta con las coordenadas geográficas sobre las que se tomaron los datos, se estudiarán dos modelos que describen la fluctuación de la gravedad con el ángulo $\theta$, estos se describen más adelante.

#Detección del ruido
El método para detectar el ruido, fue el siguiente: analizando algunas de las gráficas se pudo ver que el ruido generaba picos y no un movimiento parabólico. Por tanto, se evaluaron la cantidad de picos, es decir la cantidad de veces que un valor para $y$ era menor que el vecino de su izquierda y el de su derecha. ya que esto en una parábola perfecta no pasa, es una buena manera de comparar los datos buenos del ruido. Pero debido a que los datos buenos no son parábolas totalmente perfectas, entonces solo se eliminaron los datos que contaban con más de 4 picos, esto en pseudo-código se ve como:

peaks =0

for all $y_i$

$\quad$ if $y_{i-1}$>$y_i$ AND $y_{i+1}$>$y_i$

$\quad \quad$ peaks = peaks +1

if (peaks>4)

$\quad$ noise!

Mediante este proceso de detección se encontró que habían 884 archivos con datos con datos válidos y los otros eran ruido

A continuación se muestran dos gráficas de y contra t para unos datos correctos, junto con una con unos datos malos.

#Análisis para g contra $\theta$

Por otro para $\theta$ se observa una fuerte dependencia del ángulo con la magnitud de la gravedad. Por tanto vamos a hacer un ajuste de mínimos cuadrados para los dos modelos mostrados en la guía. Para esto primero queremos ver cómo se comporta la función $F(\theta)$

$$
F(\theta) = 1- \frac{g(\theta)}{\left< g\right>}
$$

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.scatter(theta, g,color='g', alpha=0.5, s=5)
ax.set_title(r"Gravedad contra $\theta$")
ax.set_xlabel(r"$\theta$",fontsize=16)
ax.set_ylabel("$g$",fontsize=16)
filename = 'Gravedad contra theta'
savefig("Plots/" + filename + '.pdf',format = 'pdf', transparent=True)
```

```python
prom = sum(g)/len(g)
def F(g):
    return 1 - g/prom
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.scatter(theta, F(g),color='g', alpha=0.5, s=5)
ax.set_title(r"$F(\theta)$ contra $\theta$")
ax.set_xlabel(r"$\theta$",fontsize=16)
ax.set_ylabel(r"$F(\theta)$",fontsize=16)
filename = 'F contra theta'
savefig("Plots/" + filename + '.pdf',format = 'pdf', transparent=True)
```

#Ajuste de mínimos cuadrados
Se hará un ajuste de mínimos cuadrados para dos modelos diferéntes, en el primero se supone que $F$ es de la forma

$$
F(\theta) = a_0 + a_1 cos(2\theta)
$$

Y en el segundo se supone que $F$ es de la forma

$$
F(\theta) = b_0 + b_1\theta + b_2\theta^2
$$

```python
fi = open('data_F.dat','wb')
for i in range(len(theta)):
    fi.write("%f %f \n" %(theta[i], F(g[i])))
fi.close()
```

```python
theta_sorted = copy.deepcopy(theta)
theta_sorted.sort()
```

#Primer Modelo:

```python
!cc fit_1_F.c
!./a.out
constants_1 = np.loadtxt("constants_F_data_1.dat")
print "a_0=",constants_1[0], "a_1=",constants_1[1]
```

```python
def F1(x):
    return constants_1[0] + constants_1[1]*cos(2*x*pi/180.0)
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.scatter(theta,F(g), color = 'g', alpha = 0.5, s=5,label="Experimental data")
ax.plot(theta_sorted,F1(theta_sorted), color ='b', label ="Least square (model 1)")
ax.set_title(r"$F(\theta)$ contra $\theta$")
ax.set_xlabel(r"$\theta$",fontsize=16)
ax.set_ylabel(r"$F(\theta)$",fontsize=16)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc=2)
filename = 'F modelo 1'
savefig("Plots/" + filename + '.pdf',format = 'pdf', transparent=True)
```

Para saber qué modelo es mejor, debemos calcular, la distancia en cada punto de la gráfica.

```python
dis_squared_1=(F(g)-F1(theta))**2
total = sum(dis_squared_1)
print total
scatter(theta, dis_squared_1)
```

#Segundo Modelo:

```python
!cc fit_2_F.c
!./a.out

constants_2 = np.loadtxt("constants_F_data_2.dat")
print "b_0=",constants_2[0], "b_1=",constants_2[1], "b_2=", constants_2[2]
```

```python
def F2(x):
    return constants_2[0]+constants_2[1]*x+constants_2[2]*x**2
```

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.scatter(theta,F(g), color = 'g', alpha = 0.5, s=5,label="Experimental data")
ax.plot(theta_sorted,F2(theta_sorted),color='r', label="Least square (method 2)")
ax.set_title(r"$F(\theta)$ contra $\theta$")
ax.set_xlabel(r"$\theta$",fontsize=16)
ax.set_ylabel(r"$F(\theta)$",fontsize=16)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc=2)
filename = 'F modelo 2'
savefig("Plots/" + filename + '.pdf',format = 'pdf', transparent=True)
```

```python
dis_squared_2=(F(g)-F2(theta))**2
total = sum(dis_squared_2)
print total
scatter(theta, dis_squared_2)
```

#Comparación:
Analisando estos dos resultados podemos ver que que el mejor modelo es el primero, que dice que $F(\theta) = a_0 + a_1 cos(2\theta)$ lo que se puede saber observando las gráficas ya que la primera gráfica se ve mucho más similar a los datos experimentales. Por otro lado también se puede ver que las distancias cuadradas son casi el doble en el segundo modelo que en el primero, por lo que también podríamos concluir que el primer modelo es mejor.
En conclución el modelo más acertado es el primero, por lo que la gravedad, la podríamos escribir como:
$$
g=(1-F)\left< g\right> = (1-a_0+a_1cos(2\theta))\left< g\right>
$$
Luego si hacemos una gráfica de la gravedad, encontrada experimentalmente, con la que predice este modelo, tenemos lo siguiente:

```python
fig, ax= plt.subplots(1, 1, figsize=(8, 5))

ax.scatter(theta,g, color = 'g', alpha = 0.5, s=5,label="Experimental data")
ax.plot(theta_sorted,(1-F1(theta_sorted))*prom,color='r', label="Least square (method 1)")
ax.set_title(r"$g(\theta)$ contra $\theta$")
ax.set_xlabel(r"$\theta$",fontsize=16)
ax.set_ylabel(r"$g(\theta)$",fontsize=16)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, loc=3)
filename = 'gravedad modelo 1'
savefig("Plots/" + filename + '.pdf',format = 'pdf', transparent=True)
```
