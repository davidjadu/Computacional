```python
%pylab inline
from scipy.fftpack import fftfreq, fft
```

```python
R=1.0
a_1=0.5
a_2=1.0
a_3=-2
b_1=0.7
b_2=0.3
b_3=-2
points=100
theta = linspace(0, 2*pi, points)
```

```python
x_1=R*cos(theta)+a_1
y_1=R*sin(theta)+b_1
x_2=R*cos(theta)+a_2
y_2=R*sin(theta)+b_2
x_3=R*cos(theta)+a_3
y_3=R*sin(theta)+b_3
x_random=6.0*R*random.random(10)-3.0*R
y_random=6.0*R*random.random(10)-3.0*R
```

```python
x=concatenate((x_1,x_2,x_3,x_random))
y=concatenate((y_1,y_2,y_3,y_random))
x_not_random = concatenate((x_1,x_2))
y_not_random = concatenate((y_1,y_2))
scatter(x_not_random,y_not_random)
```

```python
h=0.01
points=ceil(6.0/h)+1
intensity=zeros((points,points))
print len(x_not_random),len(y_not_random)
```

```python
for i in range(len(x_not_random)):
    intensity[ceil((y_not_random[i]+3.0)/h),ceil((x_not_random[i]+3.0)/h)]=1.0
    intensity[(int)((y_not_random[i]+3.0)/h),(int)((x_not_random[i]+3.0)/h)]=1.0
```

```python
imshow(intensity, origin='lower')
colorbar()
```

```python
intensity_fft = fft2(intensity)
correlation_fft = intensity_fft*intensity_fft
correlation = ifft2(correlation_fft)
```
