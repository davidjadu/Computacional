```python
h=0.01
points=ceil(6.0/h)+1
intensity=zeros((points,points))
print len(x),len(y)
```

```python
for i in range(len(x)):
    intensity[ceil((y[i]+3.0)/h),ceil((x[i]+3.0)/h)]=1.0
    intensity[(int)((y[i]+3.0)/h),(int)((x[i]+3.0)/h)]=1.0
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im = ax.imshow(intensity, origin='lower')
fig.colorbar(im, ax=ax)
```

```python
intensity_fft = fft2(intensity)
correlation_fft = intensity_fft*intensity_fft
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im = ax.imshow(abs(correlation_fft), origin='lower')
fig.colorbar(im, ax=ax)
```
