```python
intensity_fft = fft2(intensity)
correlation_fft = intensity_fft*intensity_fft.conjugate()
freq = fftshift(fftfreq(points,h))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

im = ax.imshow(log(abs(fftshift(correlation_fft))), origin='lower')
ax.set_title('log(Fourier transform)')
ax.set_xlabel('Arbitrary units')
ax.set_ylabel('Arbitrary units')
fig.colorbar(im, ax=ax)
```

```python
import numpy as np
import scipy
import scipy.ndimage as ndimage
import scipy.ndimage.filters as filters
import matplotlib.pyplot as plt

neighborhood_size = 10
threshold = 35
data=abs(correlation)


data_max = filters.maximum_filter(data, neighborhood_size)
maxima = (data == data_max)
data_min = filters.minimum_filter(data, neighborhood_size)
diff = ((data_max - data_min) > threshold)
maxima[diff == 0] = 0

labeled, num_objects = ndimage.label(maxima)
slices = ndimage.find_objects(labeled)
x, y = [], []
for dy,dx in slices:
    x_center = (dx.start + dx.stop - 1)/2
    x.append(x_center)
    y_center = (dy.start + dy.stop - 1)/2    
    y.append(y_center)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

ax.scatter(x,y,alpha=0.5, color = 'gray')
im=ax.imshow(abs(correlation),origin='lower')
fig.colorbar(im,ax=ax)
```
