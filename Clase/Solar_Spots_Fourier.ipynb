```python
data = np.loadtxt("sparse_sample_monthrg.dat")
```

```python
print data
```

```python
data = data[data[:,3]!=-99]
manchas = data[:,3]
year =data[:,0]+data[:,1]*1.0/12.0
plot(year, manchas)
```

```python
from scipy.interpolate import interp1d
f=interp1d(year,manchas, kind='linear')
```

```python
npoints = 512
x=linspace(year[0],year[-1],npoints)
```

```python
from scipy.fftpack import fft, fftfreq, ifft
manchas_fft=fft(f(x))
freq_fft=fftfreq(npoints,x[1]-x[0])
```

```python
plot(freq_fft,abs(manchas_fft))
```

```python
manchas_fft_clean = manchas_fft
manchas_fft_clean[abs(manchas_fft_clean)<4000]=0
plot (freq_fft,abs(manchas_fft_clean))
```

```python
manchas_clean = ifft(manchas_fft_clean)
plot(x, manchas_clean)
plot(year, manchas)
```
