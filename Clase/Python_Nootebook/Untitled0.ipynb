```python
def func(x):
    return cos(x)
```

```python
def func_prime(x):
    return -sin(x)
```

```python
npoints = 10000
x=linspace(0,4*pi,npoints)
fu=func(x)
plot(x,fu)
```

```python
h=x[1]-x[0]
f_prime_num1 = (fu[2:-1]-fu[0:-3])/(2*h)
f_prime_true = func_prime(x)
scatter(x[1:-2],f_prime_num1)
plot(x[1:-2],f_prime_true[1:-2])
```

```python
plot(x[1:-2], abs(f_prime_true[1:-2]-f_prime_num1)/h)
```

```python
f_prime_num2 = (fu[1:]-fu[:-1])/h
scatter(x[:-1],f_prime_num2)
plot(x[:-1],f_prime_true[:-1])
```

```python
plot(x[:-1], abs(f_prime_true[:-1]-f_prime_num2)/h)
```
