#Bayesian parameter estimation

The central aim in many branches in science is to estimate $N$ parameters describing a physical model: $\Theta=(\Theta_1, \Theta_2,\ldots,\Theta_N)$ from a set of $M$ measurements, $D=(D_1, \ldots, D_M)$ with a variance $\sigma=(\sigma_1, \ldots, \sigma_M)$. This is what we did in the linear algebra section of this course.

Here we will present another mathematical rigorous way to make this estimation. The big difference is that this time we will try to describe the *belief* we have on our model using a probabilistic approach.

Using Bayes theorem we introduce the term likelihood $L\equiv L(D|\Theta)$ which can be read as the probability of having the data points $D$ given a certain value of the parameters $\Theta$. Assiming that each data point $D_i$ is distributed as a gaussian with variance $\sigma_i$, and that the data points are mutually independent, one can show that the likelihood can be written as


$$
L \propto \prod_{i=0}^{M} \exp \left[ -\frac{1}{2} \left(\frac{D_i-D}{\sigma_i}\right)^2\right]
=\exp \left[ -\frac{1}{2} \sum_{i=0}^{M}\left(\frac{D_i-D}{\sigma_i}\right)^2\right] = \exp{\left(-\frac{1}{2}\chi^2\right)}
$$

where

$$
\chi^2 = \sum_{i=0}^{M}\left(\frac{D_i-D}{\sigma_i}\right)^2
$$

A good way to estimate the best parameters in the model is to sample the likelihood using the Metropolis-Hastings algorithm. 

Below is a detailed example of this procedure in the case of a set points modeled by a line.

```python
%pylab inline
#load the list of points
obs_data = loadtxt("obs_data.dat")
x_obs = obs_data[:,0]
y_obs = obs_data[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return exp(-chi_squared)
```

```python
def my_model(x_obs, m, b):
    return x_obs*m + b
```

```python
m_walk = empty((0)) #this is an empty list to keep all the steps
b_walk = empty((0))
l_walk = empty((0))

m_walk = append(m_walk, random.random())
b_walk = append(b_walk, random.random())

y_init = my_model(x_obs, m_walk[0], b_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print m_walk
print b_walk
print l_walk
```

```python
n_iterations = 20000 #this is the number of iterations I want to make
for i in range(n_iterations):
    m_prime = np.random.normal(m_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)

    y_init = my_model(x_obs, m_walk[i], b_walk[i])
    y_prime = my_model(x_obs, m_prime, b_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    alpha = l_prime/l_init
    if(alpha>=1.0):
        m_walk  = append(m_walk,m_prime)
        b_walk  = append(b_walk,b_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            m_walk = append(m_walk,m_prime)
            b_walk = append(b_walk,b_prime)
            l_walk = append(l_walk, l_prime)
        else:
            m_walk = append(m_walk,m_walk[i])
            b_walk = append(b_walk,b_walk[i])
            l_walk = append(l_walk, l_init)
```

```python
scatter(m_walk, b_walk)
```

```python
scatter(m_walk, -log(l_walk))
```

```python
scatter(b_walk, -log(l_walk))
```

```python
min_b = amin(b_walk)
max_b = amax(b_walk)
min_m = amin(m_walk)
max_m = amax(m_walk)
grid_m, grid_b = mgrid[min_m:max_m:200j, min_b:max_b:200j]
```

```python
from scipy.interpolate import griddata
n_points = size(m_walk)
points = ones((n_points,2))
print shape(points)
points[:,0] = m_walk
points[:,1] = b_walk
grid_l = griddata(points, -log(l_walk), (grid_m, grid_b), method='cubic')
imshow(grid_l.T, extent=(min_m,max_m,min_b,max_b), aspect='auto',origin='lower')
```

```python
count, bins, ignored =plt.hist(m_walk, 20, normed=True)
```

```python
count, bins, ignored =plt.hist(b_walk, 20, normed=True)
```

```python
max_likelihood_id = argmax(l_walk)
best_m = m_walk[max_likelihood_id]
best_b = b_walk[max_likelihood_id]
print log10(l_walk[max_likelihood_id])
print best_m
print best_b
```

```python
best_y = my_model(x_obs, best_m, best_b)
scatter(x_obs,y_obs)
plot(x_obs, best_y)
```

#Exercise

Using a MCMC method on the data in "hands_on/lin_algebra/movimiento.dat" to estimate the parameters describing the parabolic trajectory.

```python
data = np.loadtxt("movimiento.dat")
```

```python
x_obs=data[:,0]
y_obs=data[:,1]
```

```python
scatter(x_obs,y_obs)
```

```python
def model(x,a,b,c):
    return a*x**2+b*x+c
```

```python
def likelihood(y_obs, y_model):
    chi_squared = (1.0/2.0)*sum((y_obs-y_model)**2)
    return (1/chi_squared)**20 #Trying to mimic exponential behavior
```

```python
a_walk = empty((0)) 
b_walk = empty((0))
c_walk = empty((0))
l_walk = empty((0))

a_walk = append(a_walk, random.random())
b_walk = append(b_walk, random.random())
c_walk = append(c_walk, random.random())

y_init = my_model(x_obs, m_walk[0], b_walk[0])
l_walk = append(l_walk, likelihood(y_obs, y_init))
print a_walk
print b_walk
print c_walk
print l_walk
```

```python
n_iterations = 20000
for i in range(n_iterations):
    a_prime = np.random.normal(a_walk[i], 0.1) 
    b_prime = np.random.normal(b_walk[i], 0.1)
    c_prime = np.random.normal(c_walk[i], 0.1)

    y_init = model(x_obs, a_walk[i], b_walk[i],c_walk[i])
    y_prime = model(x_obs, a_prime, b_prime, c_prime)
    
    l_prime = likelihood(y_obs, y_prime)
    l_init = likelihood(y_obs, y_init)
    
    alpha = l_prime/l_init
    if(alpha>=1.0):
        a_walk  = append(a_walk,a_prime)
        b_walk  = append(b_walk,b_prime)
        c_walk = append(c_walk,c_prime)
        l_walk = append(l_walk, l_prime)
    else:
        beta = random.random()
        if(beta<=alpha):
            a_walk = append(a_walk,a_prime)
            b_walk = append(b_walk,b_prime)
            c_walk = append(c_walk,c_prime)
            l_walk = append(l_walk, l_prime)
        else:
            a_walk = append(a_walk,a_walk[i])
            b_walk = append(b_walk,b_walk[i])
            c_walk = append(c_walk,c_walk[i])
            l_walk = append(l_walk, l_init)
```

```python
scatter(a_walk,b_walk)
```

```python
count, bins, ignored =plt.hist(m_walk, 20, normed=True)
```

```python
max_likelihood_id = argmax(l_walk)
best_a = a_walk[max_likelihood_id]
best_b = b_walk[max_likelihood_id]
best_c = c_walk[max_likelihood_id]
print log10(l_walk[max_likelihood_id])
print best_a
print best_b
print best_c
```

```python
best_y = model(x_obs, best_a, best_b, best_c)
scatter(x_obs, y_obs)
plot(x_obs, best_y)
```
