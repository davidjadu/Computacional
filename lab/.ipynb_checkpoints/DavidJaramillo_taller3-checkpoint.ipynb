```python
%pylab inline
from scipy.interpolate import interp1d
```

```python
data = loadtxt("sparse_sample_monthrg.dat")
```

```python
year = data[:,0]
month = data[:,1]
deviation=data[:,3]
```

```python
year_40_original=[]
deviation_40_original=[]
n = len(year)
for i in range(n):
    if(year[i]>=1955):
        year_40_original.append(year[i]+month[i]*(1.0/12.0))
        deviation_40_original.append(deviation[i])
```

```python
scatter(year_40_original, deviation_40_original)
```

```python
indices_borrados=[]
year_40_borrado = year_40_original
deviation_40_borrado = deviation_40_original
for j in range(15):
    i=(int)(random.random()*len(year_40_borrado))
    indices_borrados.append(i)
    year_40_borrado = delete(year_40_borrado,i)
    deviation_40_borrado = delete(deviation_40_borrado,i)
```

#Interpolación Polinomial

```python
def polinomial(x,x_in,y_in):
    n=len(x_in)
    p=0
    for i in range(n):
        p_i=y_in[i]+0.0
        for j in range(n):
            if (i!=j):
                p_i*=(x-x_in[j])/(x_in[i]-x_in[j])
        p+=p_i
    return p
```

#Interpolación Lineal

```python
def lineal (x, x_in, y_in):
    n= len(x_in)
    pos = 0
    for i in range(n-1):
        if (x>=(x_in[i]) and x<=(x_in[i+1])):
            pos = i
    return y_in[pos]+(y_in[pos]-y_in[pos+1])/(x_in[pos]-x_in[pos+1])*(x-x_in[pos])
```

```python
y_lineal=[]
for j in range(len(x)):
    y_lineal.append(lineal(x[j],year_40_borrado,deviation_40_borrado))
```

```python
ax = axes()
ax.set_title("Linear")
ax.set_xlabel("Year")
ax.set_ylabel("Deviation")
plot(x, y_lineal)
scatter(year_40_borrado,deviation_40_borrado)
```

#Interpolación cúbica 

```python
f_cubic = interp1d(year_40_borrado, deviation_40_borrado, kind='cubic')
```

```python
y_cubico = f_cubic(x)
```

```python
plot(x,y_cubico)
scatter(year_40_borrado,deviation_40_borrado)
ax=axes()
ax.set_title("Interpolacion cubica")
ax.set_xlabel("Year")
ax.set_ylabel("Deviation")
```

#Comparación
Para saber cuál es la mejor aproximación

```python
chi_lineal = 0
chi_polinomial = 0
chi_cubic = 0
for k in indices_borrados:
    chi_lineal += (year_40_original[k]-lineal(x[k],year_40_borrado,deviation_40_borrado))**2 /(year_40_original[k]+0.0)
    chi_polinomial += (year_40_original[k]-polinomial(x[k],year_40_borrado,deviation_40_borrado))**2 /(year_40_original[k]+0.0)
    chi_cubic += (year_40_original[k]-f_cubic(x[k]))**2 /(year_40_original[k]+0.0)
print "chi cuadrado lineal: ", chi_lineal
print "chi cuadrado cúbico: ", chi_cubic
print "chi cuadrado polinomial: ", chi_polinomial
```

Podemos ver que mientras el polinomio nos da un error bastante grande, las otras dos interpolaciones nos dan errores mucho menores y muy parecidos entre sí. En la gráfica siguiente podemos observar las dos primeras interpolaciones con lo que las podresmos comparar.

```python
scatter(year_40_original,deviation_40_original, label="Datos Originales", color = "Blue")
scatter(year_40_borrado,deviation_40_borrado, label="Datos borrados", color= "Red", alpha = 0.5)
plot(x,y_lineal, label="Linear")
plot(x,y_cubico, label = "Cubic")
ax = axes()
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels)
ax.legend(labels)
```

#Análisis
Podemos ver que la interpolación polinomial es bastante mala, siempre retorna valores que están bastante alejados de lo que se espera, lo que también se ve en la grafica donde se muestra la interpolación polinomial. Por tanto es la primera que debería se descartada.

Por otro lado vemos que las interpolaciones lineales y cúbicas tienen un chi cuadrado relativamente pequeño (si lo comparamos con la interpolación polinomial). Luego ambas son buenas aproximaciones para el comportamiento de la desviación. Ahora bien si repetimos el experimento varias veces veremos que en algunas, el chi cuadrado lineal es menor que el chi cuadrado cúbico y algunas veces veremos lo contrario; por lo que podríamos llegar a pensar que ambas son equivalentes. Pero, esto no es cierto, sin lugar a dudas, si nos remitimos a las gráficas vemos que la interpolación cúbica tienede a tener picos entre puntos que están relativamente alejados entre sí y en algunos momentos se aleja bastante de los puntos. Lo que hace que chi cuadrado cúbico nos de pequeño es que entre esos picos (muchas veces) no hay puntos que hayan sido borrados.

En conclución aunque ambos chi-cuadrados nos dan muy similares, un análisis más completo nos diría que la interpolación lineal es muy superior a la interpolación cúbica, y por consiguiente a la polinomial también
